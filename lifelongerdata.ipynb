```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
```

```python
df = pd.read_csv("LiveLongerData.csv",sep=",")
```

```python
df = df.drop(['Sources', 'Links', 'ID'], axis=1)
df
```

```python
df.fillna("no comment")
```

```python
positive = df[df['Years gained / lost'] > 0]
positive
```

```python
negative = df[df['Years gained / lost'] < 0]
negative
```

```python
df = pd.DataFrame(positive)
positive_sorted = positive.sort_values(by='Years gained / lost')

plt.figure(figsize=(15, 6))
plt.bar(positive_sorted['Factor'], positive_sorted['Years gained / lost'], color='blue')
plt.xticks(rotation=90,fontsize=13)
plt.xlabel('Factor',fontsize=20)
plt.ylabel('Years gained / lost',fontsize=15)
plt.title('Distribution of Years gained by Factor',fontsize=20)
plt.tight_layout()

plt.show()
```

```python
df = pd.DataFrame(negative)
negative_sorted = negative.sort_values(by='Years gained / lost')
plt.figure(figsize=(15, 6))
plt.bar(negative_sorted['Factor'], negative_sorted['Years gained / lost'], color='red')
plt.xticks(rotation=90,fontsize=20)
plt.xlabel('Factor',fontsize=20)
plt.ylabel('Years gained / lost',fontsize=15)
plt.title('Distribution of years lost by Factor',fontsize=20)
plt.tight_layout()

plt.show()
```
